```python
sec["date"] = sec["datetime"].dt.date

# The day of the week with Monday=0, Sunday=6.
sec["ft_dt_dow"] = sec["datetime"].dt.dayofweek
# Weekday as a decimal number, where 0 is Sunday and 6 is Saturday.
sec["ft_dt_dow_str"] = sec["datetime"].dt.strftime("%w:%a")

# get_features_cnd(sec)
# get_features_kaggle(sec)

sec.columns
sec
```

```python
# or join in SQL?
sec_1d["date"] = sec["datetime"].dt.date
sec.merge(sec_1d, how="left", on="date", suffixes=["", "_1d"])
```

```python
sec["all"] = True

sec["entry_time"] = sec[(sec.ft_dt_dow==0) & (sec.datetime.dt.hour==15)]["all"]
sec.entry_time.fillna(False, inplace=True)
sec["exit_time"] = sec[(sec.ft_dt_dow==1) & (sec.datetime.dt.hour==18)]["all"]
# sec["exit_time"] = sec[(sec.ft_dt_dow==2) & (sec.datetime.dt.hour==20)]["all"]
sec.exit_time.fillna(False, inplace=True)

sec["signal_long"] = sec["entry_time"]
sec["signal_long"] &= (sec["close"] < sec["close"].shift(24))
# sec["signal_long"] &= (sec["close"].shift(24) < sec["close"].shift(48))
sec["signal_long_exit"] = sec["exit_time"]

sec.signal_long.sum()
```

```python
def fill_bt_columns(sec):
   if "signal_short" not in sec.columns:
      sec["signal_short"] = False
   if "signal_short_exit" not in sec.columns:
      sec["signal_short_exit"] = False
   if "signal_long_exit" not in sec.columns:
      sec["signal_long_exit"] = False
   if "signal_long" not in sec.columns:
      sec["signal_long"] = False

fill_bt_columns(sec)
```

```python
from backtest_lib_market_entry import MarketStrategy


class TurnAroundStrategy(MarketStrategy):

  def init(self):
    super().init()

    def fn_dow(): return self.data.Ft_dt_dow
    self.dow = self.I(fn_dow, overlay=False)
```

```python
from backtest_lib_stop_entry import get_bt_results
from backtest_lib_stop_entry import bt_convert_frame
import backtest_lib_stop_entry
backtest_lib_stop_entry.sec = sec

# bt = get_bt_results(sec, TurnAroundStrategy)
bt_df = bt_convert_frame(sec)
bt = Backtest(bt_df,
              TurnAroundStrategy,
              cash=10000,
              trade_on_close=True,
              exclusive_orders=True,
              commission=0.001
              )

output = bt.run()

bt._results
```
